```python
%matplotlib inline

#Solucionador de Ecuaciones Diferenciales (ED)
from scipy.integrate import odeint
import scipy.optimize as scp
#graficacion
import matplotlib.pyplot as plt  
#manejo de arrays
import numpy as np
#libreria matemática
import math
#constantes físicas
from scipy.constants import Planck
#cargamos libreria para manipular la legend
from matplotlib.font_manager import FontProperties


class schrodinger_eq_solver(object):    

    
    '''Soluciona la ecuacion de Schrodinger para tres diferentes tipos de potencial
    Por defecto se tiene la solución para un potencial de tipo oscilador armónico.
    Entre las opciones se tiene: pozo y pozo infinito'''
    
    def __init__(self, potential = 'oscilador'): 

        
        if potential == 'oscilador':
            self.routine1()
        if potential == 'pozo':
            self.routine2()
        if potential == 'pozo infinito':
            self.routine3()
    

    
    def routine1(self):
        
        # Parámetros
        m = 1
        
        # Función de potencial de oscilador armónico. Recibe un punto en el eje x y la constante
        def Oscilador(x):

            # Constante de elasticidad
            k = 2

            return (1/2)*k*x**2

        
        # Función de la ecuación de Schrödinger
        def psi(Psi, x, V):

            # Linealizando la ec. de Schrödinger
            # x_1' = x_2 = psi     
            # x_2' = (2m/hb**2)*(V(x)-E)x_1 = psi'

            return [Psi[1], (2*m)*(V(x) - E)*Psi[0]]
        
        
        
        # Condiciones de frontera, valores nulos de la función de onda en los extremos del intervalo y 
        # derivada muy cercana a cero  
        
        
        x0 = [0.0, 0.1]
    

        # Intervalo de soluciṕn 
        # N puntos
        pts = 1000
        val = np.linspace(-5,5,pts)

        print("Los autovalores de energía E_n son:\n")

        # Figura
        fig, ax = plt.subplots(1,4,figsize=(18,6))
        plt.suptitle("Funciones de onda $\psi(x)$", size = '20' )

        # Solución de la ec de Scrödinger. NO NORMALIZADA.

        #matriz para almacenar las autofunciones correspondientes a En
        aa = np.empty((0,pts))
        #matriz para almacenar los autovalores En
        energy = np.array([])

        for i in range(0,4):   

            # Autovalor de energía
            E = 2**0.5*(i+0.5)

            energy = np.append(energy,E)

            print("E_%d = "%i, E)

            # Solución de la ecuación y gráfica
            sol = odeint(psi, x0, val, args=(Oscilador,))   
            ax[i].plot(val,sol[:,0],color = "red",label = " " )

            aa = np.concatenate((aa,[sol[:,0]]))

        plt.show()

        # Figura
        fig, ax = plt.subplots(1,4,figsize=(18,6))
        plt.suptitle("Densidades de probabilidad $|\psi(x)|^2$", size = '20' )

        # Densidad de probabilidad NO NORMALIZADAS

        for i in range(0,4):    
            E = 2**0.5*(i+0.5)
            sol = odeint(psi, x0, val, args=(Oscilador,)) 
            ax[i].plot(val,sol[:,0]**2,color = "red",label = " " )
        plt.show()  
        
        print('\n')        
        print('Y considerando superposición dependencia temporal se tiene que:', end='\n')

        #Generador aleatorio de los coeficientes Cn
        t = np.random.random((1,aa.shape[0]))
        #norma de los coeficientes
        norm = (t*t).sum()
        t = t/(norm**0.5)
        #sum coeficientes ==1, si desea comprobar
        # print((t*t).sum())



        #matriz diagonal cuyos autovalores estan dados por los términos exponenciales de las energías E_n
        def mat_E(En,time):
            h = Planck
            #Matriz identidad de tamaño 4x4
            A = np.identity(4)
            for i in range(0,len(En)):
                n_complex = -1j
                #considerando la parte real de las exponenciales
                A[i] = A[i]*np.exp(n_complex*time*En[i]/h).real
            return A


        fontP = FontProperties()

        #arreglo de tiempos
        time = np.linspace(0,2,5)

        #graficación
        fig, ax = plt.subplots(2,1,figsize=(8,6))
        plt.subplots_adjust(hspace=0.5)


        for k in time:
            #matriz diagonal con autovalores exponenciales de En
            En = mat_E(energy,k)
            #multiplicamos el Cada coeficiente exp con sus respectivos valores de autofunciones
            A = np.matmul(En,aa)
            #multiplicamos los coeficientes generados con la matriz anterio
            wf = np.matmul(t,A)    
            plt.xlabel(r'$x$')
            ax[0].plot(val,(wf.T), label='time {:.1f}'.format(k))
            ax[0].set_title("Función de onda ${\Psi (x,t)}$", size=15)
            ax[1].plot(val,(wf.T)**2, label='time {:.1f}'.format(k))
            ax[1].set_title(r"Densidad de probabilidad $|\psi(x,t)|^2$" ,size=15)
            ax[0].legend(title='Arrays', bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP)
            
    def routine2(self):
        
        # Parámetros
        m = 1
        #Ancho & profundidad
        l = 2
        V_0 = 4
        
        # Recibe un valor del eje x, el ancho tal que el potencial es diferente de cero en [-l,l] 
        # y la profundidad del pozo V_0
        def Pozo(x):

            # Ancho y profundida del pozo 
            l = 2
            V_0 = 4

            if x<-l or x>l:
                return V_0
            if x>=-l and x<=l:
                return 0

        
        # Función de la ecuación de Schrödinger
        def psi(Psi, x, V):

            # Linealizando la ec. de Schrödinger
            # x_1' = x_2 = psi     
            # x_2' = (2m/hb**2)*(V(x)-E)x_1 = psi'

            return [Psi[1], (2*m)*(V(x) - E)*Psi[0]]
        


        # Funciones de energías para las funciones de onda simétricas y antisimétrica
        def E_sym(E):
            return np.sqrt(V_0-E) - np.sqrt(E)*np.tan(np.sqrt(E/2)*2*l) 

        def E_asym(E):
            return np.sqrt(V_0-E) + np.sqrt(E)*1/np.tan(np.sqrt(E/2)*2*l) 


        #Esta parte se utiliza para hallar las de energías más facilmente usando guess de la raíz
        guess = np.round(np.arange(0.1, 4, 0.1),6)

        # Arreglo de autovalores de energía
        ENER = []

        # Se depuran las raíces
        for i in range(0,len(guess)):

            # Raíces usando el arreglo guess como valor cercano a la raíz real
            a = np.round(scp.fsolve(E_sym, guess[i])[0], 6)
            b = np.round(scp.fsolve(E_asym, guess[i])[0], 6)

            # El algoritmo a veces retorna el valor del guess por tanto se descartan estos
            if a not in guess:
                ENER.append(a)
                #print(a)

            if b not in guess:
                ENER.append(b)
                #print(b)

        # Se quitan los valores repetidos
        ENER = np.unique(ENER)
        # Se organizan de menor a mayor
        ENER = np.sort(ENER)
        # Se toman solo los 4 primeros
        ENER = ENER[:4]
        
        
        # Condiciones iniciales e intervalo de solución
        x0 = [0.0, 0.1]
        pts = 1000
        val = np.linspace(-5,5,pts)

        print("Los autovalores de energía E_n son:\n")

        # Figura
        fig, ax = plt.subplots(1,4,figsize=(18,6))
        plt.suptitle("Funciones de onda $\psi(x)$", size = '20' )

        # Solución de la ec de Scrödinger. NO NORMALIZADA

        #matriz para almacenar las autofunciones correspondientes a En
        aa = np.empty((0,pts))
        #matriz para almacenar los autovalores En
        energy = np.array([])

        for i in range(0,4):

            #Valores de energía

            E = ENER[i]
            energy = np.append(energy,E)        
            print("E_%d = "%i, E)

            # Solución y gráfica
            sol = odeint(psi, x0, val, args=(Pozo,)) 
            ax[i].plot(val,sol[:,0],color = "red",label = " " )
            aa = np.concatenate((aa,[sol[:,0]]))

        plt.show() 


        # Figura
        fig, ax = plt.subplots(1,4,figsize=(18,6))
        plt.suptitle("Densidades de probabilidad $|\psi(x)|^2$", size = '20' )

        # Densidad de probabilidad. NO NORMALIZADA.
        for i in range(0,4):    

            E = ENER[i]

            sol = odeint(psi, x0, val, args=(Pozo,)) 
            ax[i].plot(val,sol[:,0]**2,color = "red",label = " " )
        plt.show()
        
        #Generador aleatorio de los coeficientes Cn
        t = np.random.random((1,aa.shape[0]))
        #norma de los coeficientes
        norm = (t*t).sum()
        t = t/(norm**0.5)
        #sum coeficientes ==1, si desea comprobar
        # print((t*t).sum())

        #matriz diagonal cuyos autovalores estan dados por los términos exponenciales de las energías E_n
        def mat_E(En,time):
            h = Planck
            #Matriz identidad de tamaño 4x4
            A = np.identity(4)
            for i in range(0,len(En)):
                n_complex = -1j
                #considerando la parte real de las exponenciales
                A[i] = A[i]*np.exp(n_complex*time*En[i]/h).real
            return A

        fontP = FontProperties()

        #arreglo de tiempos
        time = np.linspace(0,2,5)

        #graficación
        fig, ax = plt.subplots(2,1,figsize=(8,6))
        plt.subplots_adjust(hspace=0.5)


        for k in time:
            #matriz diagonal con autovalores exponenciales de En
            En = mat_E(energy,k)
            #multiplicamos el Cada coeficiente exp con sus respectivos valores de autofunciones
            A = np.matmul(En,aa)
            #multiplicamos los coeficientes generados con la matriz anterio
            wf = np.matmul(t,A)    
            plt.xlabel(r'$x$')
            ax[0].plot(val,(wf.T), label='time {:.1f}'.format(k))
            ax[0].set_title("Función de onda ${\Psi (x,t)}$", size=15)
            ax[1].plot(val,(wf.T)**2, label='time {:.1f}'.format(k))
            ax[1].set_title(r"Densidad de probabilidad $|\psi(x,t)|^2$" ,size=15)
            ax[0].legend(title='Arrays', bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP)
        

        
        
    def routine3(self):
        
        # Parámetros
        m = 1
        
        # Recibe un valor del eje x, el ancho tal que el potencial es diferente de cero en [-l,l] 
        def PozoInf(x):

            # Ancho y profundida del pozo 
            l = 5.0
            V_0 = 1e100

            if x<-l or x>l:
                return V_0
            if x>=-l and x<=l:
                return 0

        
        # Función de la ecuación de Schrödinger
        def psi(Psi, x, V):

            # Linealizando la ec. de Schrödinger
            # x_1' = x_2 = psi     
            # x_2' = (2m/hb**2)*(V(x)-E)x_1 = psi'

            return [Psi[1], (2*m)*(V(x) - E)*Psi[0]]
        
        # Condiciones iniciales e intervalo de solución
        x0 = [0.0, 0.1]
        pts = 1000
        val = np.linspace(-5,5,pts)
        l = 5.0

        print("Los autovalores de energía E_n son:\n")

        # Figura
        fig, ax = plt.subplots(1,4,figsize=(18,6))
        plt.suptitle("Funciones de onda $\psi(x)$", size = '20' )

        # Solución de la ec de Scrödinger. NO NORMALIZADA
        #matriz para almacenar las autofunciones correspondientes a En
        aa = np.empty((0,pts))
        #matriz para almacenar los autovalores En
        energy = np.array([])


        for i in range(0,4):

            #Valores de energía

            E = ((i+1)**2*np.pi**2)/(8*l**2)
            energy = np.append(energy,E)

            print("E_%d = "%(i+1), E)

            # Solución y gráfica
            sol = odeint(psi, x0, val, args=(PozoInf,)) 
            ax[i].plot(val,sol[:,0],color = "red",label = " " )
            aa = np.concatenate((aa,[sol[:,0]]))

        plt.show() 


        # Figura
        fig, ax = plt.subplots(1,4,figsize=(18,6))
        plt.suptitle("Densidades de probabilidad $|\psi(x)|^2$", size = '20' )

        # Densidad de probabilidad. NO NORMALIZADA.
        for i in range(0,4):    

            E = ((i+1)**2*np.pi**2)/(8*l**2)

            sol = odeint(psi, x0, val, args=(PozoInf,)) 
            ax[i].plot(val,sol[:,0]**2,color = "red",label = " " )
        plt.show()
        
        #Generador aleatorio de los coeficientes Cn
        t = np.random.random((1,aa.shape[0]))
        #norma de los coeficientes
        norm = (t*t).sum()
        t = t/(norm**0.5)
        #sum coeficientes ==1, si desea comprobar
        # print((t*t).sum())

        #matriz diagonal cuyos autovalores estan dados por los términos exponenciales de las energías E_n
        def mat_E(En,time):
            h = Planck
            #Matriz identidad de tamaño 4x4
            A = np.identity(4)
            for i in range(0,len(En)):
                n_complex = -1j
                #considerando la parte real de las exponenciales
                A[i] = A[i]*np.exp(n_complex*time*En[i]/h).real
            return A

        
        fontP = FontProperties()

        #arreglo de tiempos
        time = np.linspace(0,2,5)

        #graficación
        fig, ax = plt.subplots(2,1,figsize=(8,6))
        plt.subplots_adjust(hspace=0.5)


        for k in time:
            #matriz diagonal con autovalores exponenciales de En
            En = mat_E(energy,k)
            #multiplicamos el Cada coeficiente exp con sus respectivos valores de autofunciones
            A = np.matmul(En,aa)
            #multiplicamos los coeficientes generados con la matriz anterio
            wf = np.matmul(t,A)    
            plt.xlabel(r'$x$')
            ax[0].plot(val,(wf.T), label='time {:.1f}'.format(k))
            ax[0].set_title("Función de onda ${\Psi (x,t)}$", size=15)
            ax[1].plot(val,(wf.T)**2, label='time {:.1f}'.format(k))
            ax[1].set_title(r"Densidad de probabilidad $|\psi(x,t)|^2$" ,size=15)
            ax[0].legend(title='Arrays', bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP)
```
